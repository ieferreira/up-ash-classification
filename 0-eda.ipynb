```python
import matplotlib.pyplot as plt
import numpy as np
import cv2 
```

```python
def graphImgs(imgs, title=None, cmap="Greys_r", figsize=(12,6)):
    """ imgs ->  a list of images
        title ->  a list of strings        
        returns mpl subplots
    """
    fig, axes = plt.subplots(nrows=1, ncols=len(imgs), figsize=figsize)
    for i, ax in enumerate(axes):
        ax.imshow(imgs[i], cmap=cmap)
        if title is not None:
            try:
                ax.set_title(title[i])
            except:
                pass
    fig.tight_layout()

def average_image(folder):
    """make an average image of all image contained in folder"""
    images = os.listdir(folder)
    average_image = np.zeros(img.shape)
    for image in images:
        average_image += cv2.imread(folder + "/" + image)
    average_image = average_image / len(images)
    return average_image/255


plt.imshow(average_image("blocky"))
```

```python
graphImgs([average_image("blocky"), average_image("elongated"), average_image("rounded"), average_image("vesicular")], ["Blocky", "Elongated", "Rounded", "Vesicular"])
```
